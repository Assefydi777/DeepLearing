```python
from unsloth import FastLanguageModel
import torch

model_name = "unsloth/GLM-4.5-Air"

max_seq_length = 2048  # Choose sequence length
dtype = None  # Auto detection

# Load model and tokenizer
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name=model_name,
    max_seq_length=max_seq_length,
    dtype=dtype,
    load_in_4bit=True,
)
```

```python
# Add LoRA adapters
model = FastLanguageModel.get_peft_model(
    model,
    r=64,  # LoRA rank - higher = more capacity, more memory
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj",
    ],
    lora_alpha=128,  # LoRA scaling factor (usually 2x rank)
    lora_dropout=0,  # Supports any, but = 0 is optimized
    bias="none",     # Supports any, but = "none" is optimized
    use_gradient_checkpointing="unsloth",  # Unsloth's optimized version
    random_state=3407,
    use_rslora=False,  # Rank stabilized LoRA
    loftq_config=None, # LoftQ
)
```

```python
import json
import random
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load the dataset (replace with actual path if different)
with open("Sleep_health_and_lifestyle_dataset.json") as f:
    data = json.load(f)

# Choose a random entry with a sleep disorder (for better test value)
sample = random.choice([entry for entry in data if entry.get("Sleep Disorder") != "None"])

# Format user prompt
user_prompt = f"""Predict the most likely type of sleep disorder and explain your reasoning:

Gender: {sample['Gender']}
Age: {sample['Age']}
Occupation: {sample['Occupation']}
Sleep Duration: {sample['Sleep Duration']} hours
Quality of Sleep: {sample['Quality of Sleep']}
Physical Activity Level: {sample['Physical Activity Level']}
Stress Level: {sample['Stress Level']}
BMI Category: {sample['BMI Category']}
Blood Pressure: {sample['Blood Pressure']}
Heart Rate: {sample['Heart Rate']} bpm
Daily Steps: {sample['Daily Steps']}
"""

# Ground truth
ground_truth = sample["Sleep Disorder"]

# Chat message format
messages = [
    {
        "role": "user",
        "content": user_prompt
    }
]

# Initialize model + tokenizer (adjust as needed)
model_path = "THUDM/chatglm3-6b"  # replace with your local path or model repo
tokenizer = AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_path, trust_remote_code=True).half().cuda()
model = model.eval()

# Enable fast inference
from utils import FastLanguageModel  # make sure you have this or similar utility
FastLanguageModel.for_inference(model)

# Tokenize
inputs = tokenizer.apply_chat_template(
    messages,
    tokenize=True,
    add_generation_prompt=True,
    return_tensors="pt"
).to("cuda")

# Generate response
outputs = model.generate(
    input_ids=inputs,
    max_new_tokens=256,
    use_cache=True,
    temperature=0.7,
    do_sample=True,
    top_p=0.9,
)

# Decode and print
response = tokenizer.batch_decode(outputs, skip_special_tokens=True)[0]
print("=== Prompt ===")
print(user_prompt)
print("\n=== Model Response ===")
print(response)
print("\n=== Ground Truth ===")
print("Sleep Disorder:", ground_truth)
```
